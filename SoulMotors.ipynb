```python
#Variáveis importantes - Serviços
carrinhodeCompra = []
listavalores = []
somaCarrinho = 0
trocadeoleo = 40
trocadepneu = 35
alinhamento = 100 
balanceamento = 120
injecaoeletronica = 800
servicosdemecanica = 500
outrosservicos = 50


#Variáveis importantes - Produtos
calota = 20
lampadaFarol = 15
litroOleo = 45
limpadorParaBrisa = 10
capaVolante = 35
somautomotivo = 500





flagMenuOpcoes = True
flagPrincipal = True
print("Bem vindo(a) à Oficina Soul Motors!")
while flagPrincipal == True:
  ent = str(input("Digite uma das opções abaixo:\n\t--- A: Acessar o programa\n\t--- F: Sair do programa\n")).upper()
  if ent !="A" and ent != "F":
    print("Digite uma opção válida!")
  elif ent == "F":
    print("Saindo...")
    print("A Oficina Soul Motors agradece pela preferência!\n")
    flagPrincipal = False
  elif ent == "A":
    flagPrincipal = False
    while flagMenuOpcoes == True:
      print("Bem-vindo ao Menu de opções!")
      opcao = int(input("Digite a opção a ser realizada:\n\t1 --- Serviços\n\t2 --- Produtos\n\t3 --- Editar pedido\n\t4 --- Finalizar compra\n\t5 --- Voltar ao Menu Principal\nSua opção: "))
      if opcao<1 or opcao >5:
        print("Opção inválida. Tente novamente!")

# Entrada para o menu de seleção de serviços
      elif opcao == 1:
        flagMenuOpcoes = False
        flagServicos = True
        while flagServicos ==  True:
          print("---------------------Serviços------------------")
          print("1 - Troca de óleo ................... R$ 40,00")
          print("2 - Troca de Pneu individual......... R$ 35,00")
          print("3 - Alinhamento...................... R$ 100,00")
          print("4 - Balanceamento ................... R$ 120,00")
          print("5 - Serviços de Injeção Eletrônica... R$ 800,00")
          print("6 - Serviços de Mecânica (motor)..... R$ 500,00")
          print("7 - Voltar ao menu de opções")      
          
          entrada = int(input("Digite o numero do serviço desejado ou 7 para voltar ao menu de opções: "))  
          if entrada == 1:
            quantidade = int(input('Quantidade: '))
            carrinhodeCompra.append('troca de oleo')
            somaCarrinho += trocadeoleo*quantidade
            listavalores.append(trocadeoleo*quantidade)
            print(f'Troca de óleo adicionada. Quantidade: {quantidade}\n')
          elif entrada == 2:
            quantidade = int(input("Quantidade: "))
            carrinhodeCompra.append('troca de pneu')
            somaCarrinho += trocadepneu*quantidade
            listavalores.append(trocadepneu*quantidade)
            print(f'Troca de pneu adicionada. Quantidade: {quantidade}\n')
          elif entrada ==3:
            quantidade = int(input("Quantidade: "))
            carrinhodeCompra.append('Alinhamento')
            somaCarrinho += alinhamento*quantidade
            listavalores.append(alinhamento*quantidade)
            print(f'Alinhamento adicionado. Quantidade: {quantidade}\n')
          elif entrada == 4:
            quantidade = int(input('Quantidade: '))
            carrinhodeCompra.append('Balanceamento')
            somaCarrinho += balanceamento*quantidade
            listavalores.append(balanceamento*quantidade)
            print(f'Balanceamento adicionado. Quantidade: {quantidade}')
          elif entrada ==5:
            quantidade = int(input('Quantidade: '))
            carrinhodeCompra.append('Injeção Eletronica')
            somaCarrinho += injecaoeletronica*quantidade
            listavalores.append(servicosdemecanica*quantidade)
            print(f"Injeção eletrônica adicionada. Quantidade: {quantidade}")
          elif entrada ==6:
            quantidade = int(input('Quantidade: '))
            carrinhodeCompra.append('Servicos de Mecanica')
            somaCarrinho += servicosdemecanica*quantidade
            listavalores.append(servicosdemecanica*quantidade)
            print(f"Serviços de mecânica adicionado. Quantidade: {quantidade}")            
          elif entrada ==7:
            flagServicos = False
            flagMenuOpcoes = True

#Entrada para o menu de seleção de produtos
      elif opcao == 2:
        flagMenuOpcoes = False
        flagProdutos = True
        while flagProdutos ==  True:
          print("Digite o numero de 1 produto desejado por vez ou 7 para retornar ao menu de opções: ")
          print("1 - Calota ........................R$20,00")
          print("2 - Lâmpada do farol...............R$15,00")
          print("3 - Litro do Oleo..................R$45,00")
          print("4 - Limpador de Para-brisa.........R$10,00")
          print("5 - Capa para o volante ...........R$35,00")
          print("6 - Som automotivo.................R$500,00")
          print("7 - Voltar ao menu de opções")
          acessorios = int(input("Digite o número do produto desejado ou 7 para voltar ao menu de opções: "))
         
          if acessorios == 1:
            quantidade = int(input("Qantidade: "))
            carrinhodeCompra.append('Calota')
            somaCarrinho += calota*quantidade
            listavalores.append(calota*quantidade)
            print(f"Calota adicionada. Quantidade: {quantidade}")
          elif acessorios == 2:
            quantidade = int(input("Qantidade: "))
            carrinhodeCompra.append('Lâmpada do Farol')
            somaCarrinho += lampadaFarol*quantidade
            listavalores.append(lampadaFarol*quantidade)
            print(f"Lâmpada do Farol adicionada. Quantidade: {quantidade}")
          elif acessorios == 3:
            quantidade = int(input("Qantidade: "))
            carrinhodeCompra.append('Litro do Óleo')
            somaCarrinho += litroOleo*quantidade
            listavalores.append(litroOleo*quantidade)
            print(f"Litro do Óleo adicionado. Quantidade: {quantidade}")
          elif acessorios == 4:
            quantidade = int(input("Qantidade: "))
            carrinhodeCompra.append('lLimpador de Para-brisa')
            somaCarrinho += limpadorParaBrisa*quantidade
            listavalores.append(limpadorParaBrisa*quantidade)
            print(f"Limpador de para-brisa adicionado. Quantidade: {quantidade}")
          elif acessorios == 5:
            quantidade = int(input("Qantidade: "))
            carrinhodeCompra.append('Capa do Volante')
            somaCarrinho += capaVolante*quantidade
            listavalores.append(capaVolante*quantidade)
            print(f"Capa do Volante adicionada. Quantidade: {quantidade}")
          elif acessorios == 6:
            quantidade = int(input("Qantidade: "))
            carrinhodeCompra.append('Som Automotivo')
            somaCarrinho += somautomotivo*quantidade
            listavalores.append(somautomotivo*quantidade)
            print(f"Som Automotivo selecionado. Quantidade: {quantidade}")
          elif acessorios == 7:
            flagProdutos = False
            flagMenuOpcoes = True
  
    
#Entrada para o menu de edição de pedido
      elif opcao == 3:
        flagMenuOpcoes = False
        flagEdit = True
        while flagEdit == True:
          print("----------------- Editar pedido ---------------")
          print("\t1 - Visualizar itens\n\t2 - Remover itens\n\t3 - Voltar ao menu principal")
          entrada = int(input("Digite sua opção: "))
          if entrada == 1:
            for itens in carrinhodeCompra:
              print(itens)
          
            print(f"Total: {somaCarrinho} ")
          elif entrada == 2:
            for (i, itens) in enumerate(carrinhodeCompra, start=0):
              print(i, itens)   
            opcao = int(input("Digite o indice dos itens a remover: "))
              
            carrinhodeCompra.pop(opcao)
            somaCarrinho = somaCarrinho-listavalores[opcao]
          elif entrada == 3:
            flagEdit = False
            flagMenuOpcoes = True

#Entrada para o menu de pagamento            
      elif opcao == 4:
        flagPagamento = True
        total = somaCarrinho
        print("---------------- Finalizar Compra -----------------")
        print(f"Este é seu carrinho de compra: {carrinhodeCompra}")
        print (f'O valor total do seu pedido é R$ {somaCarrinho}.')

        fechamento = int(input("Deseja finalizar sua compra?\n\t--- Digite 1 para 'sim' e 2 para 'não'\n "))
        if fechamento == 1:
          while flagPagamento == True:
            pg = int(input("Informe a forma de pagamento desejada:\n1 - Dinheiro\n2 - Cartão\n"))
            pg < 1 and pg > 2

            if pg == 1:
              print("Você escolheu dinheiro.")
              flagPagamento = False
              flagMoney = True
              
              while flagMoney == True:
                valor = int(input(f"Levando em consideração que o total de sua compra é {somaCarrinho}, o valor pago em dinheiro é: R$"))
                valor < total
                print("Valor pago inferior ao total de compra. Por favor, digite um valor válido!")

                if valor >= total:
                  flagMoney = False
                  troco = int(valor-total)
                  
                  cinquenta = int(troco/50)
                  troco = troco - (cinquenta*50)

                  vinte = int(troco/20)
                  troco = troco - (vinte*20)

                  dez = int(troco/10)
                  troco = troco - (dez*10)

                  cinco = int(troco/5)
                  troco = troco - (cinco*5)

                  dois = int(troco/2)
                  troco = troco - (dois*2)

                  um = int(troco/1)
                  troco = troco - (um*5)

                  troco = valor - total
                  print(f'Seu troco é de R${troco}')

                  print("-------- Quantidade de notas a serem recebidas ---------")                      
                  print('Notas R$ 50,00 = ',cinquenta)
                  print('Notas R$ 20,00 = ',vinte)
                  print('Notas R$ 10,00 = ',dez)
                  print('Notas R$  5,00 = ',cinco)
                  print('Notas R$  2,00 = ',dois)
                  print('Notas R$  1,00 = ',um)
                  print(f"Você adquriu: {carrinhodeCompra}, totalizando: R${somaCarrinho}\nA Oficina Soul Motors agradece pela preferência!\n")
                  carrinhodeCompra.clear()
                  listavalores.clear()
                  somaCarrinho = 0
                      
                  flagMenuOpcoes = False
                  flagPrincipal = True

            if pg == 2:
              flagCartao = True
              print("Você escolheu pagamento via cartão.")
              while flagCartao == True:
                c = int(input("Informe a opção desejada:\n1 - Crédito\n2 - Débito\nEscolha: "))
                if c<1 or c>2:
                  flagCartao = True
                  flagPagamento = False
                  
                elif c == 1:
                  flagCredito = True
                  flagCartao = False
                  flagPagamento = False
                  print("Você escolheu cartão de crédito.")
                  while flagCredito == True:
                    parc = int(input("Digite a quantidade de parcelas entre 1 e 12\nParcelas: "))
                    if parc >= 1 and parc <= 12:
                      flagCredito = False
                      print(f'Sua compra de R$ {total} foi dividida no cartão de crédito em {parc} parcela(s) de R$ {(total/parc):.2f}.\nCompra aprovada.')
                    

                      print(f"Você adquriu: {carrinhodeCompra}, totalizando: R${somaCarrinho}\nA Oficina Soul Motors agradece pela preferência!\n")
                      carrinhodeCompra.clear()
                      listavalores.clear()
                      somaCarrinho = 0
                      flagMenuOpcoes = False
                      flagPrincipal = True

                    
                elif c == (2):
                  flagCartao = False
                  flagPagamento = False
                  print("Você escolheu cartão de débito.")
                  print(f"Sua compra de R$ {total} foi efetuada em débito!")
                  print(f"Você adquriu: {carrinhodeCompra}, totalizando: R${somaCarrinho}\nA Oficina Soul Motors agradece pela preferência!\n")
                  carrinhodeCompra.clear()
                  listavalores.clear()
                  somaCarrinho = 0
                  flagMenuOpcoes = False
                  flagPrincipal = True

    
      
                  
        elif fechamento == 2:
          print("Você tem certeza que deseja desistir da sua compra?")
          for n in carrinhodeCompra:
            print(n)
          print(f"Total: {somaCarrinho}")
          confirmacao = int(input("Digite 1 para prosseguir e desistir, ou 2 para retornar ao menu principal\n"))
          if confirmacao == 1:         
            carrinhodeCompra.clear()
            listavalores.clear()
            somaCarrinho = 0
            print("A Oficina Soul Motors agradece pela preferência!\n")
            flagMenuOpcoes = False
            flagPrincipal = True        
          


          elif confirmacao == 2:
            flagMenuOpcoes = True

#Entrada para a opção de saída do Menu de opções
      elif opcao == 5:
        flagMenuOpcoes = False
        flagPrincipal = True
    flagMenuOpcoes = True

        

          
          



```
