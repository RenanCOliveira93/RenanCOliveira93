```python
# CONSTRUIR MINHA SPARK SESSION 

spark = (
    SparkSession.builder
                .master('local')
                .appName('projeto_individual')
                .config('spark.ui.port', '4050')
                .getOrCreate()
)

display(spark)
```

```python
# COMO VOU CRIAR UM STRUCTTYPE, PRECISO SABER O TIPO DE CADA COLUNA E SE ELAS SÃO NULL ABLE

df3.printSchema()
```

```python
# CONSTRUÇÃO DO MEU STRUCT TYPE

schema = StructType([
		StructField('ano_nascimento', IntegerType(), True),
		StructField('escolaridade', StringType(), True),
		StructField('estado_civil', StringType(), True),
    StructField('renda_anual', IntegerType(), True),
    StructField('num_criancas_casa', IntegerType(), True),
    StructField('num_adolescentes_casa', IntegerType(), True),
    StructField('cliente_desde', StringType(), True),
    StructField('dias_ultima_compra', IntegerType(), True),
    StructField('gasto_com_vinho_2ultimos_anos', IntegerType(), True),
    StructField('gasto_com_frutas_2ultimos_anos', IntegerType(), True),
    StructField('gasto_com_carnes_2ultimos_anos', IntegerType(), True),
    StructField('gasto_com_peixes_2ultimos_anos', IntegerType(), True),
    StructField('gasto_com_doces_2ultimos_anos', IntegerType(), True),
    StructField('gasto_com_ouro_2ultimos_anos', IntegerType(), True),
    StructField('qtd_descontos_feitos', IntegerType(), True),
    StructField('compras_pela_internet', IntegerType(), True),
    StructField('compras_usando_catalago', IntegerType(), True),
    StructField('compras_feitas_lojaFisica', IntegerType(), True),
    StructField('visitantes_site_mensal', IntegerType(), True),
    StructField('reclamacao_2anos', IntegerType(), True)
]
)
```

```python
# DECIDI RENOMEAR 7 COLUNAS POR OUTRO NOME COM MAIS SENTIDO A COLUNA, PRA ISSO
# CRIEI OUTRO DATAFRAE COM AS COLUNAS RENOMEADAS
df3 = df3.withColumnRenamed('gasto_com_vinho_2ultimos_anos', 'Compra_vinho_2anos')
df3 = df3.withColumnRenamed('gasto_com_frutas_2ultimos_anos', 'Compra_frutas_2anos')
df3 = df3.withColumnRenamed('gasto_com_carnes_2ultimos_anos', 'Compra_carnes_2anos')
df3 = df3.withColumnRenamed('gasto_com_peixes_2ultimos_anos', 'Compra_peixes_2anos')
df3 = df3.withColumnRenamed('gasto_com_doces_2ultimos_anos', 'Compra_doces_2anos')
df3 = df3.withColumnRenamed('gasto_com_ouro_2ultimos_anos', 'Compra_ouro_2anos')
df3 = df3.withColumnRenamed('reclamacao_2anos', 'Reclamou_de_algo')

df3.show()
```

CRIAÇÃO DE COLUNAS

```python
# VOU CRIAR UMA COLUNA DE IDADE PRA SABER ALGUNS DADOS REFERENTES A IDADES DOS CLIENTES

df3 = df3.withColumn("idade", 2022 - F.col("ano_nascimento"))
```

```python
# UMA COLUNA COM O SOMATÓRIO DE CRIANÇAS E ADOLESCENTES PRA SABER A QUANTIDADE DE FILHOS

df3 = df3.withColumn('n_Filhos', F.col('num_criancas_casa') + F.col('num_adolescentes_casa'))
```

```python
# UMA COLUNA COM O TOTAL DE COMPRAS REALIZADAS
df3 = df3.withColumn('Total_compras', F.col('Compra_vinho_2anos') + F.col('Compra_frutas_2anos') + F.col('Compra_carnes_2anos') + F.col('Compra_peixes_2anos') + F.col('Compra_doces_2anos') + F.col('Compra_ouro_2anos'))
```

```python
# UMA COLUNA COM O SOMATÓRIO DE VENDAS REALIZADAS PARA VER A PORCENTAGEM DE COMPRAS FEITAS EM CADA TIPO (INTERNET, LOJA, CATALAGO)
df3 = df3.withColumn('Total_vendas', F.col('compras_pela_internet') + F.col('compras_usando_catalago') + F.col('compras_feitas_lojaFisica'))
```

```python
df3.show()
```

INSIGHTS E AGRUPAMENTOS




```python
# QUAL NÍVEL DE ESCOLARIDADE POSSUI MAIOR NUMERO DE FILHOS E QUAL COMPRAM MAIS?
df3.groupBy(F.col("escolaridade")).sum("n_Filhos", "Total_compras").show()


'''
Podemos ver que pessoas com pelo menos o nível Bacharel tem mais filhos, enquanto
que o nivel educacional fundamental apresenta um numero de filhos muito menor.
O mesmo se observa na quantidade de itens comprados em 2 itens, pessoas com Bacharel
ja fizeram quase 700 mil compras, enquanto pessoas de nivem mais basico não chegou nem a 
4500 compras realizadas durante 2 anos
'''
```

```python
# QUAL ESTADO CIVIL FEZ MAIS RECLAMAÇÕES NOS ULTIMOS 2 ANOS?

df3.groupBy(F.col("estado_civil")).sum("Reclamou_de_algo").show()

'''
Como apresenta a tabela, gente casada fazem mais reclamações do que outros
estados relacionais.
Puxar a atenção para os estados civís Sozinho, Assexuados, Viuvos e Livres,
que não prestaram nenhuma reclamação no mesmo periodo
'''
```

```python
# QUEM SÃO OS IDOSOS SEM FILHOS DESSA TABELA?
df3.filter((F.col("idade") >=70) & (F.col("n_Filhos") == 0)).show(50)

'''
Pelo que se vê nas 50 primeiras linhas, a maioria possui doutorado ou mestrado,
são casadas ou estão em união estável e  compram muito mais vinho do que carne, peixes, doces e ouro.

'''
```

```python
# QUAL GRUPO DE ESCOLARIDADE COMPRAm MAIS COM DESCONTO?

df3.groupBy(F.col("escolaridade")).sum("Total_compras", "qtd_descontos_feitos").show()

'''
Mais uma vez podemos perceber que o grupo Bacharel se destaca mais, sendo 
a escolaridade que mais compram com desconto.
'''
```

WINDOW FUNCTIONS

```python
# VOU CRIAR UMA WINDOW FUNCTION COM BASE NO TOTAL DE VENDAS PARA CADA GRAU DE ESCOLARIDADE

windowFunc  = Window.partitionBy("escolaridade").orderBy("Total_vendas")
```

```python
# ROW NUMBER PRA VIR UMA CLASSIFICAÇÃO DO PRIMEIRO AO ULTIMO SEPARADO PELOS GRUPOS

df3.withColumn("row_number", F.row_number().over(windowFunc)).show()
```

```python
# NESTA WINDOW FUNCTION, ESTOU TRAZENDO A IDADE MÉDIA DO GRUPO

df3 = df3.withColumn("media", avg(F.col("idade")).over(windowFunc))
df3.show()
```
